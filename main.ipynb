```python
import pandas as pd
```

```python
movies_df = pd.read_csv('data/movies.csv')
movies_df.shape
```

```python
movies_df.head()
```

```python
credits_df = pd.read_csv('data/credits.csv')
credits_df.shape
```

```python
credits_df.head()
```

```python
movies_df = movies_df.merge(credits_df, on='title')
movies_df.shape
```

```python
movies_df.info()
```

```python
movies_df = movies_df[['movie_id', 'title', 'genres', 'overview', 'keywords', 'cast', 'crew']]
movies_df.shape
```

```python
movies_df.info()
```

```python
movies_df.isna().sum()
```

```python
movies_df.dropna(inplace=True)
```

```python
movies_df.duplicated().sum()
```

```python
movies_df.shape
```

```python
import ast
def get_all_name(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_df['genres'] = movies_df['genres'].apply(get_all_name)
movies_df['keywords'] = movies_df['keywords'].apply(get_all_name)
```

```python
movies_df.head()
```

```python
def get_3_name(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L
```

```python
movies_df['cast'] = movies_df['cast'].apply(get_3_name)
```

```python
movies_df.head()
```

```python
def get_director_name(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies_df['crew'] = movies_df['crew'].apply(get_director_name)
```

```python
movies_df.head()
```

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x: x.split())
```

```python
movies_df.head()
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x: [i.replace(' ', '') for i in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x: [i.replace(' ', '') for i in x])
movies_df['cast'] = movies_df['cast'].apply(lambda x: [i.replace(' ', '') for i in x])
movies_df['crew'] = movies_df['crew'].apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
movies_df.head()
```

```python
movies_df['tags'] = movies_df['genres'] + movies_df['overview'] + movies_df['keywords'] + movies_df['cast'] + movies_df['crew']
```

```python
movies_df.head()
```

```python
res_movies_df = movies_df[['movie_id', 'title', 'tags']]
```

```python
res_movies_df.head()
```

```python
res_movies_df['tags'] = res_movies_df['tags'].apply(lambda x: ' '.join(x))
res_movies_df['tags'] = res_movies_df['tags'].apply(lambda x: x.lower())
```

```python
res_movies_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
res_movies_df['tags'] = res_movies_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000, stop_words='english')
```

```python
vectors = cv.fit_transform(res_movies_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = res_movies_df[res_movies_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(res_movies_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Iron Man')
```

```python
recommend('Captain America: Civil War')
```
